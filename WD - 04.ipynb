```python
import matplotlib.pyplot as plt
import collections
from collections import Counter
import seaborn as sns
```

```python
#https://wolnelektury.pl/media/book/txt/stevenson-dr-jekyll-i-mr-hyde.txt
```

```python
f = open("stevenson-dr-jekyll-i-mr-hyde.txt", 'r', encoding='utf-8')
text=f.read().split()
f.close()
```

```python
slowa=list(text)
slowa2=slowa
```

```python
znaki_interpunkcyjne=['.', ',', '!', '?', '(', ')', '"','"', ':']
for i in (0, len(znaki_interpunkcyjne)-1):
    slowa2=list(map(lambda x: x.replace(znaki_interpunkcyjne[i], ''), slowa2))
```

```python
slowa3=list(map(lambda x: x.lower(), slowa2))
```

```python
len(slowa3)
```

```python
slowa4=[]
for i in range (0, len(slowa3)):
    if len(slowa3[i])>=5:
        slowa4.append(slowa3[i])
```

```python
slowa5=Counter(slowa4)
slowa6=list(slowa5.most_common(20))
slowa7=dict(slowa6)
slowa7
```

```python
fig=plt.figure(figsize=(15,12))
ax=fig.add_subplot()

plt.barh(range(len(slowa7)), list(slowa7.values()), align='center', tick_label=list(slowa7.keys()), color=sns.color_palette("Blues", len(slowa7)))

for i, v in enumerate(list(slowa7.values())):
    ax.text(v/2 , i , str(v), color='black', fontweight='bold', size=12)

plt.title('20 najczęściej występujących słów(pięcioliterowych lub dłuższych)', size=20)
ax.set_xlabel("Ilość", size=15)

plt.savefig("04_mac_273926.svg")
plt.show()
```
