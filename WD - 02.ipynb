```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x1=random.uniform(0,2)
x2=random.uniform(2,4)
x3=random.uniform(4,6)
x4=random.uniform(6,8)
```

```python
def f(x):
    return (x-x1)*(x-x2)*(x-x3)*(x-x4)
```

```python
N=100000
x=np.linspace(-1,8,N)
y=f(x)
dydx=np.gradient(y,x)
dydx[1:0]
dydx[:-1]
ext=x[1:][dydx[1:]*dydx[:-1]<0]
ext
```

```python
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot()

ax.plot(np.linspace(-1,ext[0],N),f(np.linspace(-1,ext[0],N)), color="purple", linewidth=5)
ax.plot(np.linspace(ext[0],ext[1],N), f(np.linspace(ext[0],ext[1],N)), color="red", linewidth=5)
ax.plot(np.linspace(ext[1],ext[2],N),f(np.linspace(ext[1],ext[2],N)), color="purple", linewidth=5)
ax.plot(np.linspace(ext[2], 10, N), f(np.linspace(ext[2], 10,N)), color="red", linewidth=5)

ax.scatter(ext, f(ext), marker=".", color="black", s=800, zorder=3)
ax.set_ylim(-50,50)

ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_position(('data', 0))

ax.annotate(round(f(ext[1])-f(ext[0]),4), xy=((ext[0]+ext[1])/2, (f(ext[0])+f(ext[1]))/2), 
            xytext=((ext[0]+ext[1])/2, -30),arrowprops=dict(color='green', width=1), fontsize=30)
ax.annotate(round(f(ext[2])-f(ext[1]),4), xy=((ext[1]+ext[2])/2, (f(ext[1])+f(ext[2]))/2), 
            xytext=((ext[1]+ext[2])/2, -30),arrowprops=dict(color='green', width=1), fontsize=30)

plt.title(f"Losowe pierwiastki: {round(x1,4), round(x2,4), round(x3,4), round(x4,4)}", fontsize = 40)

plt.savefig("02_mac_273926.pdf")
```
