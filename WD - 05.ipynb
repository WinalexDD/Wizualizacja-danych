```python
import matplotlib.pyplot as plt
import numpy as np
import random
import sympy
```

```python
def f(x):
    return 20/(3*x**2+2)
```

```python
n=random.randint(1,30)
N=1000000
x=np.linspace(0,5,N)
y=f(x)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,sharex=True,sharey=True)

ax1.set_xlim(0,5)
ax1.plot(x,y, color="black")
ax2.plot(x,y, color="black")
ax3.plot(x,y, color="black")
ax2.fill_between(x,0,y, color="lightblue")
ax2.annotate(r'$f(x)=\frac {20} {3x^2+2}$', [1,8], size=14)

P1=[]
P2=[]
for i in range(0,n+1):
    ax1.scatter(i*(5/n), f(i*(5/n)), color="red", zorder=5)
    ax2.scatter(i*(5/n), f(i*(5/n)), color="red", zorder=5)
    ax3.scatter(i*(5/n), f(i*(5/n)), color="red", zorder=5)
    ax1.bar(i*(5/n), height=f(i*(5/n)), align='edge', width=5/n, color="lightblue")
    ax3.bar(i*(5/n), height=f(i*(5/n)), align='edge', width=-(5/n), color="lightblue")
    P1.append((5/n)*f(i*(5/n)))
    P2.append((5/n)*f(i*(5/n)))
    
b = sympy.Symbol("b")
POLE=sympy.integrate(f(b), (b, 0, 5))
POLE1=sum(P1)-P1[-1]
POLE2=sum(P2)-P2[0]

plt.suptitle("Przedzialow jest " + str(n) + ", pole= " + str(round(float(POLE), 6)), size=15)
ax1.set_title("suma I = " + str(round(POLE1,6)), size=9)
ax3.set_title("suma II = " + str(round(POLE2,6)), size=9)

plt.savefig("05_mac_273926.pdf")
plt.show()
```
