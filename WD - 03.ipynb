```python
import matplotlib.pyplot as plt
from math import log
import numpy as np
```

```python
T=[88,224.7,365.2,687,4331,10747,30589,59800] 
R=[57.9,108.2,149.6,227.9,778.6,1433.5,2872.5,4495.1] 
C=["yellow", "orange","blue", "grey","darkorange", "gold","lightskyblue"'Navy']
N=["Merkury", "Wenus", "Ziemia", "Mars", "Jowisz", "Saturn", "Uran", "Neptun"]
S=[1,3,4,2,8,7,5,6] #wielkosc posegregowana od najmniejszej do najwiekszej

x = [2*log(num) for num in T]
y = [3*log(num) for num in R]
```

```python
plt.style.use("dark_background")
fig=plt.figure(figsize=(15, 10))
ax=fig.add_subplot()

ax.plot(x,y)

ax.annotate(r'$\frac {T^2} {R^3} =const$', [13,22], size=50)

plt.xlabel('okres orbitalny (2*log(T), T->dni)', size=20)
plt.ylabel("polos wielka " r'$(3*log(R), R->km*10^6$)', size=20)

for n in range(0,8):
    ax.scatter (x[n], y[n], zorder=5, label=N[n], s=200*S[n])
    
    
ax.legend(prop={'size': 28})

plt.savefig("01_mac_273926.pdf")
```
