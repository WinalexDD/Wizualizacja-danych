```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*x**5-4*x**4+2*x**2-7*x
```

```python
N=100000
x=np.linspace(-1.5,2,N)
y=f(x)
plt.plot(x,y)
plt.show()
```

```python
dydx=np.gradient(y,x)
dydx[1:0]
dydx[:-1]
ext=x[1:][dydx[1:]*dydx[:-1]<0]
ext
```

```python
fig=plt.figure()
ax=fig.add_subplot()

N=100000
x=np.linspace(-1.5,2,N)
y=f(x)
ax.plot(x,y, label=(r"$3x^5-4x^4+2x^2-7x$"), color="purple")

ax.legend()
ax.set_xticks(ext)

ax.scatter(ext, f(ext), marker=".", color="black", s=200)
ax.set_ylim(-15,15)

plt.grid(True)
plt.savefig("01_mac_273926.pdf")
plt.show()
```
